# Feature Engineering

```python
# Age Group Binning using quantile range method
quantile_list = [0, .2, .4, .6, .8, 1.]
quantiles = churn_data['Age'].quantile(quantile_list)
labels = [f"{int(quantiles.iloc[i])}-{int(quantiles.iloc[i+1])}" for i in range(len(quantiles)-1)]
churn_data['AgeGroup'] = pd.qcut(churn_data['Age'], q=quantile_list, labels=labels)
```

```python
# Tenture Grouping 
tenure_quantiles = churn_data['Tenure'].quantile(quantile_list)
labels = [f"{int(tenure_quantiles.iloc[i])}-{int(tenure_quantiles.iloc[i+1])}" for i in range(len(tenure_quantiles) - 1)]
churn_data['TenureGroup'] = pd.qcut(churn_data['Tenure'], q=quantile_list, labels=labels)
```

```python
# Balance Categories
Balance_median = churn_data['Balance'].median()
churn_data['HighBalance'] = churn_data['Balance'].apply(lambda x: 1 if x > Balance_median else 0)
```

```python
# Balance and number of products
churn_data['Balance_NumOfProducts'] = churn_data['Balance'] * churn_data['NumOfProducts']
```

```python
# Encoding categorical features
churn_data = pd.get_dummies(churn_data, columns=['Geography', 'Gender'], drop_first=True)
```

```python
churn_data.head()
```

# Data Pre-Processing

```python
# Removing Unnecessary Columns
# Drop the 'Customer ID' and 'Surname' columns as they are not useful for modeling
churn_data.drop(['CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
# Scaling Numerical Features

scaler = StandardScaler()
```

```python
# List of numerical columns that need scaling
numerical_columns = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary', 'Balance_NumOfProducts']
# Apply scaling to the numerical columns
churn_data[numerical_columns] = scaler.fit_transform(churn_data[numerical_columns])
```

```python
# Encoding Categorical Variables
# Encode 'AgeGroup' and 'TenureGroup' using Label Encoding
label_encoder = LabelEncoder()
churn_data['AgeGroup'] = label_encoder.fit_transform(churn_data['AgeGroup'])
churn_data['TenureGroup'] = label_encoder.fit_transform(churn_data['TenureGroup'])
```

# Handling Class Imbalance in the Dataset

```python
churn_data.head()
```

```python
# Installing imblearn package

from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Defining features (X) and target variable (y)
X = churn_data.drop('Exited', axis=1)
y = churn_data['Exited']

# Creating an instance of SMOTE
smote = SMOTE(random_state=42)

# Applying SMOTE to generate synthetic samples for balancing the classes
X_resampled, y_resampled = smote.fit_resample(X, y)

# Splitting the Dataset into Training and Testing Sets
X_train, X_val, y_train, y_val = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)
```

```python
print(f"X_train: {X_train.shape}\ny_train: {y_train.shape}\nX_val: {X_val.shape}\ny_val:{y_val.shape}")
```

# Training and Evaluating Model Performance¶

```python
!pip install xgboost
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Initialize an XGBoost classifier
xgb_clf = xgb.XGBClassifier(objective='binary:logistic', eval_metric='auc', use_label_encoder=False, random_state=42)

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'colsample_bytree': [0.3, 0.7]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=xgb_clf, param_grid=param_grid, scoring='roc_auc', cv=3, verbose=2)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)
```

```python
# Best parameters and score
print("Best parameters found: ", grid_search.best_params_)
print("Best AUC found: ", grid_search.best_score_)

# Predict on the test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_val)

# Evaluation
print(classification_report(y_val, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_pred))
print("AUC-ROC:", roc_auc_score(y_val, best_model.predict_proba(X_val)[:, 1]))
```

```python
from sklearn.metrics import roc_curve, auc
# Compute ROC curve and AUC
fpr, tpr, _ = roc_curve(y_val, y_pred)
roc_auc = auc(fpr, tpr)

# Plotting ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_val, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```
